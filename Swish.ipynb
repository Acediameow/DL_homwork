```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sigmoid函数
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# swish函数
def swish(x, beta):
    return x * sigmoid(beta * x)
```

```python
# 获取直角坐标系
def get_central():
    t = plt.gca()
    t.spines['right'].set_color('none') 
    t.spines['top'].set_color('none') 
    t.xaxis.set_ticks_position('bottom') 
    t.yaxis.set_ticks_position('left') 
    t.spines['bottom'].set_position(('data', 0))
    t.spines['left'].set_position(('data', 0))
    
    return t
```

```python
# 画图
x = np.arange(-5, 5)
t = get_central() 

legends = []
for beta in [0, 0.5, 1, 100]:
    y_s = swish(x, beta)
    t.plot(x, y_s)
    legends.append('β = ' + str(beta))

t.legend(legends)
plt.show()
```
